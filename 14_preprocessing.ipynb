**preprocessing is done because the data should be understable to model , efficient , common data type (using data-transformation)** 

1.  *Normalizer* :  used to normalise the data  between some specific range , shorter the range the efficient the data is  , using the min and max values of data  

       **x***new* = (**x***a* - **x***min*)/(**x***max* - **x***min*)

2.  *Standardization* :  also called  standardisation using the standard deviation and mean , standardising the data between the specific range 

3.  *Robust scaler* : also called as InterQuartile Range(IQR) , d/nce between the 1st and  3rd outlier  , new data can be gained using  Robust scaler "\n" 
             **x***new*   =  (x-**x***median*)/(IQR)

4.  *outlier* : if the plot is outside the cluster or group of points then it is called as outlier , so detection of outlier is necessary  

5.  *Normaliser*:     applied row-wise  to get n number of values with intervals in between the values, whereas  normalisation    
    np.random.normal()


6. *Binariser* : converts continue values to categorical binary values 0&1
          Binariser(threshold) . ... the value <=threshold is converted to 0 else, 1


7.  *Encoding* : model mostly understand number values so most of the nominal , ordinal , catelgorical values are converted to the number values like 0,1,2....  like using OneHotEncoder & Label Encoder
