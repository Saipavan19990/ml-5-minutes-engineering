```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
table = pd.DataFrame(iris.data,columns=iris.feature_names)
table.head()
```

```python
table["target"] = iris.target
table.head()
```

```python
table.head()
```

```python
setosa = table.loc[table["target"]==0]
verginica = table.loc[table["target"]==1]
versicolor = table.loc[table["target"]==2]
```

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
table = pd.DataFrame(iris.data,columns=iris.feature_names)
table.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(setosa["sepal length (cm)"],np.zeros_like(setosa["sepal length (cm)"]),"o")
plt.plot(verginica["petal length (cm)"],np.zeros_like(verginica["petal length (cm)"]),"d",c="b")
plt.plot(versicolor["petal length (cm)"],np.zeros_like(versicolor["petal length (cm)"]),"s",c="y")
plt.xlabel("petal length")
plt.title("petal length using univariate")


# sepal length is the common attribute in all 3 species
```

if petal length is >6.0 then it is versicolor
if petal lenth is <4.0 then it is verginica
